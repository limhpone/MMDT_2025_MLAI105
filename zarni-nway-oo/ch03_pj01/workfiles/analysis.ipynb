```python
import pandas as pd
```

```python
df = pd.read_csv("./results/mmculture/mmculture_predictions.csv")
```

```python
df.head()
```

```python
model_name = ['ResNet50', 'VGGNet16', 'InceptionV3', 'ConvNeXt', 'EfficientNet']
```

```python
pred_classes = []
for name in model_name:
    pred_classes.append(name + "_top1")
    pred_classes.append(name + "_top2")
    pred_classes.append(name + "_top3")
```

```python
df_pred_classes = df[["label"] + pred_classes]
```

```python
df_resnet50_pred_classes = df_pred_classes[['label', 'ResNet50_top1', 'ResNet50_top2', 'ResNet50_top3']]
df_vggnet16_pred_classes = df_pred_classes[['label', 'VGGNet16_top1', 'VGGNet16_top2', 'VGGNet16_top3']]
df_inceptionv3_pred_classes = df_pred_classes[['label', 'InceptionV3_top1', 'InceptionV3_top2', 'InceptionV3_top3']]
df_convnext_pred_classes = df_pred_classes[['label', 'ConvNeXt_top1', 'ConvNeXt_top2', 'ConvNeXt_top3']]
df_efficientnet_pred_classes = df_pred_classes[['label', 'EfficientNet_top1', 'EfficientNet_top2', 'EfficientNet_top3']]
```

```python
df_resnet50_pred_classes.to_csv('./results/resnet50_pred_class.csv', index=False)
df_vggnet16_pred_classes.to_csv('./results/vggnet16_pred_class.csv', index=False)
df_inceptionv3_pred_classes.to_csv('./results/inceptionv3_pred_class.csv', index=False)
df_convnext_pred_classes.to_csv('./results/convnext_pred_class.csv', index=False)
df_efficientnet_pred_classes.to_csv('./results/efficientnet_pred_class.csv', index=False)
```
